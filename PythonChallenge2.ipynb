```python
#Data Cleaning & Feature Engineering
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.head() 
```

```python
#Drop Columns That Won't Be Useful
df.drop(columns = ['Name', 'Ticket', 'Cabin'], inplace = True)
```

```python
#Check For Missing Values
df.isnull().sum()
```

```python
#Handle Missing Value
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True) 
```

```python
# Convert categorical columns to numerical using Label Encoding
label_encoder = LabelEncoder()
df['Sex'] = label_encoder.fit_transform(df['Sex'])  # Male = 1, Female = 0
df['Embarked'] = label_encoder.fit_transform(df['Embarked'])  # C, Q, S → 0, 1, 2
```

```python
# Feature Engineering, Creating A New Feature(Column)
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
df.drop(columns = ['PassengerId'])
```

```python
# Normalise Numerical Features
scaler = StandardScaler()
df[['Age', 'Fare', 'FamilySize']] = scaler.fit_transform(df[['Age', 'Fare', 'FamilySize']])
```

```python
# Save cleaned dataset Desktop
df.to_csv("titanic_cleaned.csv", index=False)
```
